```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
df = pd.read_csv('/content/churn_modeling1.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(columns=['CustomerId', 'RowNumber', 'Surname'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.Age.min()
```

```python
df.Age.max()
```

```python
labels = ['0-20', '21-40', '41-60', '61+']
bins = [0, 20, 40, 60, 100]
```

```python
df['Age_Labels'] = pd.cut(df['Age'], bins=bins, labels=labels, include_lowest=True)
```

```python
df.head()
```

```python
df[['Age', 'Age Labels']].to_csv('age_labels.csv', index=False)
```

```python
df.Age_Labels.value_counts()
```

```python
plt.bar(labels, df.Age_Labels.value_counts())
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Customers')
add_labels(labels, df.Age_Labels.value_counts())
plt.show()
```

```python
def add_labels(x, y):
  for i in range(len(x)):
    plt.text(i, y[i], y[i], ha='center')
```
